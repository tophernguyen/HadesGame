```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("output.csv"))
```

```python
df.head()
```

```python
traits = df['trait (sub-traits)'].str.split(', ', expand=True)
```

```python
traits
```

```python
new = pd.concat([df,traits], axis=1)
```

```python
list(new.columns)
```

```python
new_column_names = {int(i): f'trait {i+1}' for i in range(31)}
new.rename(columns=new_column_names, inplace = True)
```

```python
import re
# Function to split trait and level
def split_trait(trait):
    if pd.isna(trait):
        return pd.Series([None, None])
    match = re.match(r'(.+?) \((Level \d+)\)', trait)
    if match:
        return pd.Series([match.group(1), match.group(2)])
    return pd.Series([trait, None])

# Expand each trait column
for i in range(1, 32):
    trait_col = f'trait {i}'
    new[[f'{trait_col} name', f'{trait_col} level']] = new[trait_col].apply(split_trait)
    new.drop(columns=[trait_col], inplace=True)
```

```python
new.to_csv("hades.csv", index = False)
```
